```python
#IMPORT LIBRARIES
import pandas as pd


#IMPORT DATA
csv_data = pd.read_csv("Resources/purchase_data.csv", encoding="utf-8")
df = pd.DataFrame(csv_data)
```

```python
#Preview Data
df
```

```python
#PLAYER COUNT
total_player = len(df["SN"].unique())

df_player_count = pd.DataFrame({"Total Players": [total_player]})
df_player_count
```

```python
#PURCHASING ANALYSIS (TOTAL)

#unique items
items_count = len(df["Item ID"].unique())

#average price
avg_price = '$' + str(round(df["Price"].mean(), 2))

#num of purchases
purchase_count = str(len(df["Purchase ID"]))

#total revenue
total_rev = '$' + str(df["Price"].sum())


data = {'Number of Unique Items': [items_count],
        'Average Price':[avg_price],
        'Number of Purchases':[purchase_count],
        'Total Revenue':[total_rev]}



df_purchase_analysis = pd.DataFrame(data)
df_purchase_analysis
```

```python
#COLLECT GENDER DATA
df_male = df[df.Gender.isin(['Male'])]
df_female = df[df.Gender.isin(['Female'])]
df_other = df[df.Gender.isin(['Other / Non-Disclosed'])]
```

```python
#GENDER DEMOGRAPHICS

#male
male = len(df_male["SN"].unique())
male_pct = str(round((male / total_player)*100, 2)) + '%'

#female
female = len(df_female["SN"].unique())
female_pct = str(round((female / total_player)*100, 2)) + '%'

#other
other = len(df_other["SN"].unique()) 
other_pct = str(round((other /  total_player)*100, 2)) + '%'


data = [['Male', male, male_pct], ['Female', female, female_pct], ['Other', other, other_pct]]
columns = ['Gender', 'Total Count', 'Percentage of Players']

df_gender = pd.DataFrame(data, columns = columns)
df_gender
```

```python
#Purchase Analysis Data

#Puchase Count
purch_count_female = df_female["Purchase ID"].count()

purch_count_male = df_male["Purchase ID"].count()

purch_count_other = df_other["Purchase ID"].count()


#Average Purchase Price
avg_price_female = round(df_female["Price"].mean(), 2)

avg_price_male = round(df_male["Price"].mean(), 2)

avg_price_other = round(df_other["Price"].mean(), 2)


#Total Purchase Value 
total_purch_female = round(sum(df_female["Price"]), 2)

total_purch_male = round(sum(df_male["Price"]), 2)

total_purch_other = round(sum(df_other["Price"]), 2)

#Average Total Purchase per Person
avg_purch_female = round(total_female / female, 2)

avg_purch_male = round(total_male / male, 2)

avg_purch_other = round(total_other / other, 2)
```

```python
#PURCHASING ANALYSIS (GENDER)

data = {'Gender':[' ',' ', ' ', ' '],
        'Female':[purch_count_female, '$' + str(avg_price_female), '$' + str(total_purch_female), '$' + str(avg_purch_female)],
        'Male':[purch_count_male, '$' + str(avg_price_male), '$' + str(total_purch_male), '$' + str(avg_purch_male)],
        'Other / Non-Disclosed': [purch_count_other, '$' + str(avg_price_other), '$' + str(total_purch_other), '$' + str(avg_purch_other)]
       }

columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']


df_purchase_gen = pd.DataFrame.from_dict(data, orient = 'index', columns= columns)
df_purchase_gen
```

```python
#AGE DEMOGRAPHICS

age_drop_duplicates = df.drop_duplicates(subset = "SN")

df_age = pd.DataFrame.from_dict({'Age':age_drop_duplicates["Age"]})
df_age
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_names = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']



df_age["Age Groups"] = pd.cut (df_age["Age"], bins, labels=group_names)
df_age
```

```python
age_group = df_age.groupby("Age Groups")
print(age_group["Age"].count())
```
