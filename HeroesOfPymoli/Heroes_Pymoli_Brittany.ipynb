```python
#IMPORT LIBRARIES
import pandas as pd

#IMPORT CSV
csv = pd.read_csv("Resources/purchase_data.csv", encoding = "utf-8")
df_data = pd.DataFrame(csv)
```

```python
df_data.count()
```

```python
#Preview raw dataset
df_data
```

```python
#Reduce dataset to unique SN player names only
df = df_data.drop_duplicates(subset = ['SN']).reset_index(drop=True)
df.head()
```

```python
#PLAYER COUNT

player_count = df["SN"].count()
df_total_players = pd.DataFrame({"Total Players": [player_count]})
df_total_players
```

```python
#PURCHASING ANALYSIS

#Number of unique items
unique_items = len(df_data["Item ID"].unique())

#Average Price
average_price = '$' + str(round(df_data["Price"].mean(), 2))

#Number of Purchases
purchase_count = str(df_data["Purchase ID"].count())

#Total Revenue
total_revenue = '$' + str(df_data["Price"].sum())

data = {'Unique Items':[unique_items],
        'Average Price': [average_price],
        'Total Purchases':[purchase_count],
        'Total Revenue': [total_revenue]
       }

df_purchase_analysis = pd.DataFrame(data)
df_purchase_analysis
```

```python
#GENDER DEMOGRAPHICS

#Count and Percentage of Male Players
male = df[df.Gender.isin(['Male'])]["SN"].count()
male_pct = str(round((male / player_count)*100, 2)) + '%'


#Count and Percentage of Female Players
female = df[df.Gender.isin(['Female'])]["SN"].count()
female_pct = str(round((female / player_count)*100, 2)) + '%'


#Count and Percentage
other = df[df.Gender.isin(['Other / Non-Disclosed'])]["SN"].count()
other_pct = str(round((other / player_count)*100, 2)) + '%'


#Gender Demographics Data
data = {'Gender':['Male', 'Female', 'Other'],
        'Total Count': [male, female, other],
        'Percentage of Players': [male_pct, female_pct, other_pct]
       }

columns = ['Gender', 'Total Count', 'Percentage of Players']

df_gender = pd.DataFrame(data, columns = columns)
df_gender
```

```python
#PURCHASING ANALYSIS (GENDER)

#Purchase Count
pcount_male = df_data[df_data.Gender.isin(['Male'])]["SN"].count()
pcount_female = df_data[df_data.Gender.isin(['Female'])]["SN"].count()
pcount_other = df_data[df_data.Gender.isin(['Other / Non-Disclosed'])]["SN"].count()


#Average Purchase Price
app_male = round(df_data[df_data.Gender.isin(['Male'])]["Price"].mean(), 2)
app_female = round(df_data[df_data.Gender.isin(['Female'])]["Price"].mean(), 2)
app_other = round(df_data[df_data.Gender.isin(['Other / Non-Disclosed'])]["Price"].mean(), 2)

#Total Purchase Value
tpv_male = round(df_data[df_data.Gender.isin(['Male'])]["Price"].sum(), 2)
tpv_female = round(df_data[df_data.Gender.isin(['Female'])]["Price"].sum(), 2)
tpv_other = round(df_data[df_data.Gender.isin(['Other / Non-Disclosed'])]["Price"].sum(), 2)

#Average Total Purchase per Person
atp_male = round(tpv_male / male, 2)
atp_female = round(tpv_female / female, 2)
atp_other = round(tpv_other / other, 2)

#Data Lists




data = {'Gender':["Male", 'Female', 'Other / Non-Disclosed'],
        'Purchase Count': [pcount_male, pcount_female, pcount_other],
        'Average Purchase Price': [app_male, app_female, app_other],
        'Total Purchase Value': [tpv_male, tpv_female, tpv_other],
        'Avg Total Purchase per Person': [atp_male, atp_female, atp_other]
       }

columns = ['Gender', 'Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']

df_purchase_analyis_bygender = pd.DataFrame(data, columns= columns)
df_purchase_analyis_bygender
```
